```
class SquareToCircle(Scene):
    def construct(self):
        # Configure frame
        self.camera.frame_width = 12.8  # 1280px = 12.8 units
        self.camera.frame_height = 7.2  # 720px = 7.2 units
        self.camera.background_color = WHITE

        # Solution steps and explanations
        solution_steps = [
            "Let the two numbers be \(a\) and \(b\): \(a+b=20\) and \(P= a \\times b^{2}\)",
            "\(b=20-a \\therefore P = a(20-a)^{2} = a(400-40a+a^{2}) = 400a - 40a^{2} + a^{3}\)",
            "\(P'(a)= 100 - 40a + 3a^{2}\)",
            "\(0 = (3a - 10)(a - 10)\)",
            "\(a = 10\) or \(a = \\frac{10}{3}\)",
            "If \(a = 10\): \(b = 0\) (invalid). If \(a = \\frac{10}{3}\): \(b = \\frac{20}{3}\)",
            "The numbers are \(\\frac{10}{3}\) and \(\\frac{20}{3}\)"
        ]

        explanations = [
            "We start by defining our variables and the relationship between them!",
            "Let's express everything in terms of one variable using substitution!",
            "Time for some calculus! We'll find the derivative to locate maxima!",
            "Setting derivative to zero helps find critical points!",
            "We check which solutions are valid positive numbers!",
            "Final answer! These numbers give the maximum product!"
        ]

        # Set up layout areas
        left_column = VGroup(
            *[Tex(step, color=BLACK).scale(0.45) for step in solution_steps]
        ).arrange(DOWN, aligned_edge=LEFT, buff=0.4).to_edge(LEFT, buff=1).shift(UP * 0.5)

        # Animation sequence
        current_left_content = VGroup()
        for i, (step, explanation) in enumerate(zip(left_column, explanations)):
            # Create right side content
            right_content = VGroup(
                self.create_calculations(i),
                Text(explanation, font_size=22, color=BLACK)
            ).arrange(DOWN, buff=1, aligned_edge=RIGHT).to_edge(RIGHT, buff=1).shift(UP * 0.0001)

            # Animate
            self.play(
                FadeIn(step, shift=RIGHT),
                FadeIn(right_content, shift=LEFT),
                run_time=1
            )
            current_left_content.add(step)
            self.wait(4)
            self.play(FadeOut(right_content))

        self.wait(2)

    def create_calculations(self, step_index):
        calculations = VGroup()
        if step_index == 1:
            calculations.add(MathTex("(20 - a)^2 = ", "400", "- 40a", "+ a^2", color=BLACK).scale(0.6))
            calculations.add(MathTex("P = a(400 - 40a + a^2)", color=BLACK).scale(0.6))
            calculations.add(MathTex("= 400a - 40a^2 + a^3", color=BLACK).scale(0.6))
        elif step_index == 2:
            calculations.add(MathTex("\\frac{d}{da}(400a) = 400", color=BLACK).scale(0.6))
            calculations.add(MathTex("\\frac{d}{da}(-40a^2) = -80a", color=BLACK).scale(0.6))
            calculations.add(MathTex("\\frac{d}{da}(a^3) = 3a^2", color=BLACK).scale(0.6))
            calculations.add(MathTex("P' = 400 - 80a + 3a^2", color=BLACK).scale(0.6))
        elif step_index == 3:
            calculations.add(MathTex("3a^2 - 40a + 100 = 0", color=BLACK).scale(0.6))
            calculations.add(MathTex("(3a - 10)(a - 10) = 0", color=BLACK).scale(0.6))
        elif step_index == 4:
            calculations.add(MathTex("3a - 10 = 0 \\Rightarrow a = \\frac{10}{3}", color=BLACK).scale(0.6))
            calculations.add(MathTex("a - 10 = 0 \\Rightarrow a = 10", color=BLACK).scale(0.6))
        elif step_index == 5:
            calculations.add(MathTex("b = 20 - \\frac{10}{3} = \\frac{50}{3}", color=BLACK).scale(0.6))

        return calculations.arrange(DOWN, aligned_edge=RIGHT, buff=0.3)

%manim -qm -v WARNING SquareToCircle
```
